# Análisis exploratorio de coches clásicos

En este proyecto se analizarán los datos sobre coches clásicos para hacer frente al problema presentado por la empresa.

## Definición del problema

El objetivo de la empresa Mint Classics es cerrar una de sus instalaciones de almacenamiento.

Para ello, analizaremos los datos que nos han ofrecido sobre su empresa. En base a lo que hallemos, les daremos nuestras conclusiones y propondremos opciones para llevar a cabo con el fin de estar más cerca o conseguir su objetivo.

## Conclusiones

La empresa tiene una gran cantidad de coches en stock, pero teniendo en cuenta la cantidad de coches que vende, probablemente no necesitaría contar con tanto stock. Esto se ve en que si disminuimos la cantidad de stock en un 10 % todavía está muy por encima de las ventas efectuadas.

Aún así, no hay un almacén que esté muy por debajo de los demás en cuanto a ventas o cantidad de stock.

## Recomendaciones

El objetivo a largo plazo es cerrar una de las instalaciones. Para llegar a ello, a corto plazo se recomienda bajar las existencias de stock y volver a hacer un análisis para observar los resultados.

Otra recomendación más específica es tomar estos datos es que del stock que más se puede prescindir es de los coches menos vendidos. En especial del Ford Thunderbird, que es el menos vendido y tiene más stock que otros modelos de coches.
